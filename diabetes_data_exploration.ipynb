```python
corr_matrix = explore_data.corr()

corr_matrix["Diabetes_012"].sort_values(ascending=False)
```

```python
# see what features are most correlated with one another

upper_cor_matrix = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

unique_corr = upper_cor_matrix.stack().dropna()

sorted_corr = unique_corr.abs().sort_values(ascending=False)

print(sorted_corr[:20])
# PhysHlth and GenHlth are the most correlated, folloed by DiffWalk and PhysHlth,
# and GenHlth and DiffWalk
# Education and Income are also relatively highly correlated
```
